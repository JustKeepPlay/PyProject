```python
from urllib.request import urlopen as req
from bs4 import BeautifulSoup as Soup
```

2. Request URL by get data from url https://www.tmd.go.th/province.php?id=37

```python
url = 'https://seekthailand.com/category/sneakers'
webopen = req(url)
page_html = webopen.read()
print(page_html) #report in term of bytes
webopen.close()
```

3 Convert page_html to Soup Object

```python
#you must convert bytes to web_page using beutifulsoup
data = Soup(page_html,'html.parser')
print(data)
#หา tag ที่แสดงอุณภูมิตัวองศา
#หรือหาดูได้ที่ browser โดยคลิกขวาที่อุณภูมิแล้ว inspect ดู code
```

4. Find Element (td: 'strokeme')

```python
temp = data.findAll('td',{'class':'strokeme'})
print(temp)
#ทดลองเปลี่ยนจังหวัดจาก กรุงเทพเป็นเชียงใหม่ โดยเปลี่ยนจาก id = 37  เป็น id=2
```

```python
#ตัดคำให้เหลือแค่ องศา โดยใช้ .text เพื่อเอาแค่ข้อความองศาเอา tag ออก
result = temp[0].text
print(result)
```

```python
 #ดึงค่าจังหวัดมาด้วย
 province = data.findAll('span',{'class':'title'})
 pv = province[0].text.replace(' ','') #ลบช่องว่างหน้าด้านเชียงใหม่ ที่ไม่ใช่ space แต่ให้ copy ช่องว่าด้านหน้าเชียงใหม่มาวาง
 print(pv)
 print('จังหวัด : {} อุณหภูมิ : {}'.format(pv,result))  #จัด format
 print(f'จังหวัด :{pv} อุณหภูมิ : {result}')
```

```python
#create function
def Temperature(pvid):
  url = 'https://www.tmd.go.th/province.php?id='+str(pvid)
  #pvid = provice id from website
  webopen = req(url)
  page_html = webopen.read()
  webopen.close()
  data = Soup(page_html,'html.parser')
  temp = data.findAll('td',{'class':'strokeme'}) # อุณภูมิ
  result = temp[0].text   #เหลือแค่ตัวเลข
  province = data.findAll('span',{'class':'title'}) #ชื่อจังหวัด
  pv = province[0].text.replace(' ','') #ลบช่องว่างหน้าด้านเชียงใหม่ ที่ไม่ใช่ space แต่ให้ copy ช่องว่าด้านหน้าเชี้ยงใหม่มาวาง
  #print('จังหวัด : {} อุณหภูมิ : {}'.format(pv,result))  #จัด format
  print(f'จังหวัด :{pv} อุณหภูมิ : {result}') #จัด format แบบ python v. 3
  text = 'จังหวัด : {} อุณหภูมิ : {}'.format(pv,result)
  return  text

for i in range(1,10):
  Temperature(i)

```

```python
#ดักจับ error กรณีที่ไม่มี field ที่ระบุไว้ หรือจังหวัดไม่มีข้อมูล
for i in range(25,50):
  try:
    Temperature(i)
  except Exception as e:
    print('no result because',e)
```
