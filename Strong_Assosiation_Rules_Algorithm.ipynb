# **Data Preparation**

```python
data = list(df["products"].apply(lambda x : x.split(',')))
data
```

```python
!pip install mlxtend
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_data = te.fit(data).transform(data)
df = pd.DataFrame(te_data,columns=te.columns_)
df
```

# Implementing Apriori Algorithm

```python
from mlxtend.frequent_patterns import apriori
freq_items = apriori(df,min_support=0.05,use_colnames = True) #ทดลองเปลี่ยนค่า min_support
freq_items
```

# **Create strong association rules**

```python
from mlxtend.frequent_patterns import association_rules
df_res = association_rules(freq_items, metric = "confidence", min_threshold = 0.3) #ทดลองเปลี่ยนค่า confidence or lift
df_res
```
