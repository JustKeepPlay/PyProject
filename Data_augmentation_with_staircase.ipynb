```python
import keras
print(keras.backend.backend())
import tensorflow as tf
print(tf.__version__)
import pandas as pd
from PIL import Image
import cv2
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.layers import Dense, BatchNormalization,LayerNormalization,UnitNormalization,RandomFlip,RandomRotation
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
folder_path = "/content/gdrive/MyDrive/Colab Notebooks/dataset/Staircase Dataset"
ASD_images = os.listdir(folder_path + '/50_Ascending Stairs/')
DCD_images = os.listdir(folder_path + '/50_Descending Stairs/')

dataset=[]
lab=[]

for image_name in ASD_images:
    image=cv2.imread(folder_path + '/50_Ascending Stairs/' + image_name)
    image=Image.fromarray(image,'RGB')
    image=image.resize((280,430))
    dataset.append(np.array(image))
    lab.append(0)

for image_name in DCD_images:
    image=cv2.imread(folder_path + '/50_Descending Stairs/' + image_name)
    image=Image.fromarray(image,'RGB')
    image=image.resize((280,430))
    dataset.append(np.array(image))
    lab.append(1)
```

```python
def plot_state(state):
    plt.figure(figsize= (12,12))
    for i in range(1, 10, 1):
        plt.subplot(3,3,i)
        img = load_img(folder_path + "/" + state + "/" + os.listdir(folder_path + "/" + state)[i], target_size=(240, 430))
        plt.imshow(img)
    plt.show()
```

```python
plot_state('50_Ascending Stairs')
```
