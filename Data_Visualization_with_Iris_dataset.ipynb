```python
sns.set_style('darkgrid')
g = sns.FacetGrid(iris, hue = 'species', size = 5)
g.map(plt.scatter,'petal_length','sepal_length')
g.add_legend()
plt.show()
```

```python
sns.set_style('ticks') #darkgrid
g = sns.FacetGrid(iris, hue = 'species', size = 5)
g.map(plt.scatter,'petal_length','sepal_length')
g.add_legend()
plt.show()
```

```python
# Second find relationship between Petal length and width vs Species types
pet_len_wid = iris[iris.species == 'setosa'].plot(kind = 'scatter', x = 'petal_length', y = 'petal_width'
,color = 'blue', label = 'class setosa')
iris[iris.species == 'versicolor'].plot(kind = 'scatter', x = 'petal_length', y = 'petal_width', color = 'red'
,label = 'class versicolor', ax = pet_len_wid)
iris[iris.species == 'virginica'].plot(kind = 'scatter', x = 'petal_length', y = 'petal_width', color = 'green'
,label = 'class virginica', ax = pet_len_wid)
pet_len_wid.set_xlabel('Petal Length')
pet_len_wid.set_ylabel('Petal Width')
pet_len_wid.set_title('Petal-Length-Width Vs Species')
pet_len_wid = plt.gcf()
pet_len_wid.set_size_inches(15, 7)
plt.show()
```

```python
# Scatter plots of all pairs of attributes
plt.close()
sns.pairplot(iris, hue = 'class', size = 3, diag_kind = 'kde')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1) #two row two column and show on 1
sns.violinplot(x='species',y='petal_length',data=iris)
plt.subplot(2,2,2)
sns.violinplot(x='species',y='petal_width',data=iris)
plt.subplot(2,2,3)
sns.violinplot(x='species',y='sepal_length',data=iris)
plt.subplot(2,2,4)
sns.violinplot(x='species',y='sepal_width',data=iris)
plt.show()
```
